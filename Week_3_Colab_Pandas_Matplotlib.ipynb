## ðŸŒŸ Challenge: Data Analysis

1. Create your own DataFrame with at least 4 students and their scores in 3 subjects.
2. Calculate the average score for each student.
3. Add a column called `Grade` with "Pass" or "Fail" (passing = average â‰¥ 60).


```python
# Your pandas challenge code here
import pandas as pd

# Create a DataFrame
data = {'Name': ['Alice', 'Bette', 'Carmen', 'Dana'],
        'Maths': [90, 60, 70, 80],
        'English': [90, 70, 50, 30],
        'Music': [60, 35, 90, 50]}
df = pd.DataFrame(data)

# Calculate the average for each student
df['Average'] = df[['English', 'Maths', 'Music']].mean(axis=1)
print('Average')

# Add a column called 'Grade'
df['Grade'] = df['Average'].apply(lambda x: 'Pass' if x>=60 else 'Fail')
print(df)
```

```python
import matplotlib.pyplot as plt

# Sample bar plot
names = ['Alice', 'Bob', 'Charlie']
scores = [88, 92, 79]

plt.bar(names, scores)
plt.title("Scores by Student")
plt.ylabel("Score")
plt.show()
```

## ðŸŒŸ Challenge: Plot Your Data

1. Plot your custom DataFrame (from above) using a bar plot.
2. Create a line plot showing the trend of one studentâ€™s scores across subjects.


```python
# Your plot challenge code here
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame
data = {'Name': ['Alice', 'Bette', 'Carmen', 'Dana'],
        'Maths': [90, 60, 70, 80],
        'English': [90, 70, 50, 30],
        'Music': [60, 35, 90, 50]}
df = pd.DataFrame(data)

# Calculate the average for each student
df['Average'] = df[['English', 'Maths', 'Music']].mean(axis=1)
print(df)

# Create a bar plot
# Corrected: Pass the actual data series for x and y
plt.bar(x = df['Name'], height = df['Average'])
plt.title("Average Scores by Student")
plt.xlabel("Names")
plt.ylabel("Average Scores")
plt.show()

# Create a line plot
# Corrected: Pass the actual data series for x and y
plt.plot(df['Name'], df['Average'])
plt.title("Average Scores by Student")
plt.xlabel("Names")
plt.ylabel("Average Score")
plt.show()

# Generate Descriptive Statistics (extra)
df.describe()
print(df.describe())
```
